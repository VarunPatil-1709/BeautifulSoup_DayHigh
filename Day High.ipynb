```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://ticker.finology.in/market/52-week-high"
responce=requests.get(url)
```

```python
html_doc=responce.text
```

```python
soup= BeautifulSoup(html_doc,"html.parser")
```

```python
table=soup.find("table",class_="table table-sm tickertable table-hover")
```

```python
print(soup.prettify())
```

```python
table = soup.find("table", {"data-classes": "table table-sm tickertable table-hover"})
```

```python
print(soup.table.attrs)
```

```python
table_head=table.find_all("th")
column=[]
for i in table_head:
    title=i.text
    column.append(title)
```

```python
print(column)
```

```python
table_body=table.find_all("tr")
```

```python
print(table_body)
```

```python
body=[]
for i in table_body[1:]:
    all_td=i.find_all('td')
    row=[name.text for name in all_td]
    print(row)
    body.append(row)
    
```

```python
print(body)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(body)
```

```python
df.columns=column
```

```python
df.to_csv("Day_High.csv",index=False)
```
